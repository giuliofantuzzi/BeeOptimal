## Libraries

```python
from beeoptimal import ArtificialBeeColony,Bee
from beeoptimal.benchmarks import Rastrigin2d
from beeoptimal.plotting import ContourPlotBee
import numpy as np
```

## Example code

```python
bee = Bee(position=np.array([1,1]),function=None,bounds=np.array([[-5.12,5.12],[-5.12,5.12]]))
```

```python

ABC = ArtificialBeeColony(n_bees=100,function=Rastrigin2d.fun,bounds=Rastrigin2d.bounds)
x = np.linspace(Rastrigin2d.bounds[0][0], Rastrigin2d.bounds[0][1], 100)
y = np.linspace(Rastrigin2d.bounds[1][0], Rastrigin2d.bounds[1][1], 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(x)[:, None] + np.cos(y)
points = np.c_[X.ravel(), Y.ravel()]
Z = np.array([Rastrigin2d.evaluate(p) for p in points]).reshape(X.shape)
#ContourPlotBee(x, y, Z, [bee],title='Example of ContourPlotBee')

# plot the function
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=1, fontsize=10)
plt.show()
```
