```python
from beeoptimal import ArtificialBeeColony
from beeoptimal.benchmarks import Rastrigin2d
from beeoptimal.plotting import ContourPlotBee_matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
ABC = ArtificialBeeColony(n_bees=100,function=Rastrigin2d.fun,bounds=Rastrigin2d.bounds)
```

```python
ABC.optimize(max_iters      = 1000,
             mutation       = 'ABC/best/1',
             selection      = 'RouletteWheel',
             initialization = 'random',
             verbose=True)
```

```python
x = np.linspace(Rastrigin2d.bounds[0][0]-0.05*np.abs(Rastrigin2d.bounds[0][0]), Rastrigin2d.bounds[0][1]+0.05*np.abs(Rastrigin2d.bounds[0][1]), 100)
y = np.linspace(Rastrigin2d.bounds[1][0]-0.05*np.abs(Rastrigin2d.bounds[1][0]), Rastrigin2d.bounds[1][1]+0.05*np.abs(Rastrigin2d.bounds[1][1]), 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(x)[:, None] + np.cos(y)
points = np.c_[X.ravel(), Y.ravel()]
Z = np.array([Rastrigin2d.evaluate(p) for p in points]).reshape(X.shape)
```

```python
ContourPlotBee_matplotlib(x, y, Z, ABC.colony_history[0],title='ContourPlotBee [random initialization]',optimal_solution=Rastrigin2d.optimal_solution)
```

```python
ABC.reset()
ABC.optimize(max_iters      = 1000,
             mutation       = 'ABC/best/1',
             selection      = 'RouletteWheel',
             initialization = 'cahotic',
             verbose        =True)
```

```python
ContourPlotBee_matplotlib(x, y, Z, ABC.colony_history[0],title='ContourPlotBee [cahotic initialization]',optimal_solution=Rastrigin2d.optimal_solution)
```
